```python
#Problem 1: Social Media – Post Engagement Analyzer

def post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Condition check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Input
post_engagement([200, 300, 550])
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System

def medicine_stock(stock):
    
    # Condition check
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    # Output
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example Input
medicine_stock(6)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_list, required_level):
    
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    # Calculate average
    avg = total / len(rainfall_list)
    
    # Condition check
    if avg >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Output
    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)


# Example Input
rainfall_checker([60, 70, 80, 75], 70)
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def check_duplicates(usernames):
    
    # Convert list to set
    unique_users = set(usernames)
    
    # Compare lengths
    if len(usernames) != len(unique_users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Input
check_duplicates(["user1", "user2", "user1"])
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker

def check_age(age):
    
    # Condition check
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    # Output
    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Example Input
check_age(21)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter

def premium_crops(prices):
    
    premium = []
    
    # Loop to filter premium prices
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    # Output
    print("Premium Crops:", premium)


# Example Input
premium_crops([1500, 2500, 3200, 1800])
```

```python
#Problem 7: System Monitoring – Application Health Checker

def app_health(errors):
    
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)


# Example
app_health(7)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_check(attendance_list):
    
    total_classes = len(attendance_list)
    attended = 0
    
    # Loop to count attended classes
    for a in attendance_list:
        attended += a
    
    percentage = (attended / total_classes) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example
attendance_check([1,1,1,0,1,1,1,1,0,1])
```

```python
#Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
    
    bill = 0
    
    # Billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Type:", usage)


# Example
electricity_bill(250)
```
