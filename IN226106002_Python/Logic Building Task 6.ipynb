```python
#Problem Statement 1:Smart Parking Lot Management System

def parking_system(capacity, logs):
    
    parked = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1
        
        # Track peak usage
        if parked > peak:
            peak = parked
    
    # Status check
    if parked > capacity:
        status = "Full / Over Capacity"
    else:
        status = "Available"
    
    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


# Example
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def delivery_time(distance, traffic, weather):
    
    # Base time (5 min per km)
    time = distance * 5
    
    # Traffic impact
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    
    # Weather impact
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    
    print("Estimated Delivery Time:", time, "minutes")


# Example
delivery_time(8, "High", "Rainy")
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_list):
    
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100
    
    # Status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)


# Example (simulate 150 bookings)
seat_occupancy(200, [1]*150)
```

```python
# Problem Statement 4:Cloud Server Load Classification System

def server_load(cpu_list):
    
    total = 0
    
    for cpu in cpu_list:
        total += cpu
    
    avg = total / len(cpu_list)
    
    # Classification
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("Average CPU Load:", avg, "%")
    print("Server Status:", status)


# Example
server_load([45, 60, 70, 85, 90])
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor

def resource_monitor(resources):
    
    overused = []
    
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        alert = "Yes"
    else:
        alert = "No"
    
    print("Overused Resources:", overused)
    print("Energy Alert:", alert)


# Example
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
#Problem Statement 6: Online Event Registration Capacity Controller

def event_registration(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Example
event_registration(100, 105)
```
